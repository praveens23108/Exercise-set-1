### Question 01 - Create these lists without NumPy (plain Python)

**List of 15 ones**

```python
numbers = []
for x in range (15):
    numbers.append(1)
numbers
```

**List of 15 number 7's**

```python
numbers = []
for x in range (15):
    numbers.append(7)
numbers
```

**List of numbers from 100 to 150**

```python
numbers = []
for x in range (100,151):
    numbers.append(x)
numbers
```

**List of even numbers between 0 and 100**

```python
number = []
for cycle in range(100):
    number.append(cycle * 2)
number
```

**List of years that are divisible by 4 between 1950 and 2020**

```python
start = 1952
end = 2020
for year in range(start,end,4):
    print(year)
```
