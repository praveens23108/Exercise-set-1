## Question 03. Let's try out NumPy vectors and matrices: ##

```python
# Create a NumPy vector (list) containing these values:
# 5, 16, 9, 2, 19, 18, 11, 7
# Use argmax() to find the position index of the largest value in
# the vector, and print it


number = [5, 16, 9, 2, 19, 18, 11, 7]
data = np.array(number)
number

large_value = data.argmax()
large_value
```

```python
# Create a 7x7 matrix, with values from 1 to 49
# Extra task: Create this matrix (list of lists) without NumPy in
# Python! (without the help of AI!)

data = np.arange(1,50).reshape(7,7)
data

```

```python
# Create a 7x7 matrix, with values from 1 to 49

numbers = []
for x in range (1,50):
    numbers.append(x)


for m in range(0,49,7):
    print(numbers[m:m+7])
```
